```python
# import packages
import pandas as pd
import nltk
from nltk.tokenize.casual import casual_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

# initialization
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
stop_words = nltk.corpus.stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
apps_df = pd.read_csv('googleplaystore.csv')
apps_df = apps_df[apps_df['Category'] == 'SPORTS']
names = list(apps_df['App'].unique())
print(names)
print(len(names))
```

```python
reviews_df = pd.read_csv('googleplaystore_user_reviews.csv')
display(reviews_df.head())
```

```python
reviews_df = reviews_df[reviews_df['App'].isin(names)][['App', 'Translated_Review', 'Sentiment']]
positive_reviews_df = reviews_df[reviews_df['Sentiment'] == 'Positive'][['App', 'Translated_Review']].dropna().reset_index(drop=True)
negative_reviews_df = reviews_df[reviews_df['Sentiment'] == 'Negative'][['App', 'Translated_Review']].dropna().reset_index(drop=True)
reviews_df = reviews_df[['App', 'Translated_Review']].dropna().reset_index(drop=True)

reviews_df = reviews_df.groupby('App')['Translated_Review'].apply(list).reset_index(name='Reviews')
reviews_df['Reviews'] = reviews_df['Reviews'].apply(' '.join)

positive_reviews_df = positive_reviews_df.groupby('App')['Translated_Review'].apply(list).reset_index(name='Reviews')
positive_reviews_df['Reviews'] = positive_reviews_df['Reviews'].apply(' '.join)

negative_reviews_df = negative_reviews_df.groupby('App')['Translated_Review'].apply(list).reset_index(name='Reviews')
negative_reviews_df['Reviews'] = negative_reviews_df['Reviews'].apply(' '.join)

display(reviews_df)
```

```python
def preprocess(doc):
  tokens = casual_tokenize(doc, strip_handles=True, reduce_len=True) # tokenize
  tokens = [x for x in tokens if x not in stop_words] # filter stopwords
  tokens = [str.lower(x) for x in tokens] # case folding
  tokens = [lemmatizer.lemmatize(x) for x in tokens] # lemmatizing
  doc = ' '.join(tokens)
  return doc

reviews_df['Reviews'] = reviews_df['Reviews'].apply(preprocess)
positive_reviews_df['Reviews'] = reviews_df['Reviews'].apply(preprocess)
negative_reviews_df['Reviews'] = reviews_df['Reviews'].apply(preprocess)

display(reviews_df)
```

```python
all_vectorizer = TfidfVectorizer(lowercase=True, stop_words='english')
all_vectors = all_vectorizer.fit_transform(reviews_df['Reviews'].values)
all_tfidf = pd.DataFrame(all_vectors.toarray(), columns=all_vectorizer.get_feature_names_out(), index=reviews_df['App'].values)

positive_vectorizer = TfidfVectorizer(lowercase=True, stop_words='english')
positive_vectors = positive_vectorizer.fit_transform(positive_reviews_df['Reviews'].values)
positive_tfidf = pd.DataFrame(positive_vectors.toarray(), columns=positive_vectorizer.get_feature_names_out(), index=positive_reviews_df['App'].values)

negative_vectorizer = TfidfVectorizer(lowercase=True, stop_words='english')
negative_vectors = negative_vectorizer.fit_transform(negative_reviews_df['Reviews'].values)
negative_tfidf = pd.DataFrame(negative_vectors.toarray(), columns=negative_vectorizer.get_feature_names_out(), index=negative_reviews_df['App'].values)
```

```python
keywords = dict()

for name in reviews_df['App'].values:

  k = {'all': [], 'positive': [], 'negative': []}

  s_all = all_tfidf.T[name].sort_values(ascending=False)

  try:
    s_positive = positive_tfidf.T[name]
    keywords_positive = set(s_positive[s_positive > 0].index)
  except:
    print(name + ' has no positive reviews')

  try:
    s_negative = negative_tfidf.T[name]
    keywords_negative = set(s_negative[s_negative > 0].index)
  except:
    print(name + ' has no negative reviews')

  new_keywords_negative = keywords_negative - keywords_positive
  new_keywords_positive = keywords_positive - keywords_negative

  s_positive = s_positive.filter(items=new_keywords_positive, axis=0).sort_values(ascending=False)
  s_negative = s_negative.filter(items=new_keywords_negative, axis=0).sort_values(ascending=False)

  k['all'] = list( s_all[s_all > 0].index[:20] )
  k['positive'] = list( s_positive[s_positive > 0].index[:20] )
  k['negative'] = list( s_negative[s_negative > 0].index[:20] )

  keywords[name] = k
```

```python
def getAll(name): return ' '.join(keywords[name]['all'])
def getPos(name): return ' '.join(keywords[name]['positive'])
def getNeg(name): return ' '.join(keywords[name]['negative'])

output = pd.DataFrame({'App': keywords.keys()})
output['All'] = output['App'].apply(getAll)
output['Positive'] = output['App'].apply(getPos)
output['Negative'] = output['App'].apply(getNeg)
output.head(50)
```

```python
output.to_excel('keywords.xlsx', index=False)
```
