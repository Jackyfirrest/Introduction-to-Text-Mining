```python
# 請使用者輸入一段評論，輸入評論後利用模型預測這段評論的app名稱
all_comments = []
i = 1
comment = input("please enter a comment: ")
all_comments += [comment] 
# 輸入yes以繼續輸入下一段評論
while True:
  action = input('continue? yes/no ')
  if action.lower() == 'yes':
    comment = input("please enter a comment: ")
    all_comments += [comment] 
    i = i + 1
  else:
    break
```

```python
embeddings = extract_embeddings(model_path, all_comments)
```

```python
comment_vectors = []
for num in range(0,i) : 
  comment_vectors.append(embeddings[num][0])
```

```python
comment_predict = []
comment_predict.extend(SVM_model2.predict(comment_vectors))
```

```python
from pandas import DataFrame
df = DataFrame({'comment': all_comments , 'app': comment_predict })
df.to_excel('/content/gdrive/Shareddrives/文探期末/predict.xlsx', sheet_name='sheet1', index=False)
```
