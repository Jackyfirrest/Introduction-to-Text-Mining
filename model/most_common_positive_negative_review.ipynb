```python
# We analyze sports app only
# merge two datasets
import numpy as np 
import pandas as pd
df = pd.read_csv('/content/drive/Shareddrives/文探期末/dataset/googleplaystore_user_reviews.csv')
df2 = pd.read_csv('/content/drive/Shareddrives/文探期末/dataset/googleplaystore.csv').drop_duplicates(subset = "App")
alldata = df.merge(right=df2, right_on='App', left_on='App')
alldata = alldata[alldata['Category'] == 'SPORTS']
alldata = alldata[alldata['Translated_Review'] == alldata['Translated_Review']]
alldata['App'].value_counts()
```

```python
#drop null
alldata = alldata.dropna()
```

```python
#drop negative review
negativeIndex = alldata[alldata['Sentiment_Polarity']<=0].index
positivedata = alldata.drop(negativeIndex)
positivedata.head()
```

```python
#drop positive review
positiveIndex = alldata[alldata['Sentiment_Polarity']>=0].index
negativedata = alldata.drop(positiveIndex)
negativedata.head()
```

```python
# rearrange the negativedata index
negativedata["index"] = range(0,337)
negativedata = negativedata.set_index("index")
negativedata.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
```

```python
# positive reviews preprocessing
STOPWORDS = set(stopwords.words('english'))
positivecorpus=[]
for i in range(0,938):
    review = re.sub('[^a-zA-Z]', ' ',positivedata['Translated_Review'][i])
    review = re.sub('[/(){}\[\]\|@!,;]', ' ',positivedata['Translated_Review'][i])
    review = re.sub('[^0-9a-z #+_♥️]', ' ',positivedata['Translated_Review'][i])#Remove bad symbols
    review = review.lower()
    review = review.split()
    stemmer = PorterStemmer()
    review = [stemmer.stem(token) for token in review if token not in STOPWORDS]
    review=' '.join(review)
    positivecorpus.append(review)
positivecorpus
```

```python
# positive words preprocessing
positivewords = []
Positive_Sentiment_Polarity=[]
for i in range(0,len(positivecorpus)):
    positivewords = positivewords + (re.findall(r'\w+', positivecorpus[i]))# words cantain all the words in the dataset
    Positive_Sentiment_Polarity.append
positivewords
```

```python
# count positive words 
from collections import Counter
positivewords_counts = Counter(positivewords)
```

```python
# sort positive words by their counts
most_common_positive_words = sorted(positivewords_counts.items(), key=lambda x: x[1], reverse=True)
most_commmom_positive_wordList = []
most_commmom_positive_CountList = []
for x, y in most_common_positive_words:
    most_commmom_positive_wordList.append(x)
    most_commmom_positive_CountList.append(y)
```
